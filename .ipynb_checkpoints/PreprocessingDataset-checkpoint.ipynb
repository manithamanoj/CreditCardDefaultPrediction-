```python
#Merging Dataset 1 and Dataset 2
merged_df=NewClaim_df.merge(PredictorDataset_df, how='left', on='hhld_id')
merged_df
```

```python
#Merging the merged dataset and dataset 3
merged1_df=merged_df.merge(SubsequentLossExperience_df, how='left', on='hhld_id')
merged1_df
```

```python
merged1_df=pd.DataFrame(merged1_df)
merged1_df
```

```python
#Checking whether if there is any NAN's
merged1_df.isnull().values.any()
```

```python
merged1_df.isnull().sum()
```

#### Columns 'time_w_carr','prior_bi','veh_lien_cnt','credit_score' have null values.'veh_lien_cnt'has 4293 null values and we remove that column.For rest of the three columns, we fill the Nan's with its column average.

```python
#Dropping column 'veh_lien_cnt'
new_merged_df=merged1_df.drop(['veh_lien_cnt'], axis=1)
new_merged_df
```

#### Filling Nan's

```python

new_merged_df['credit_score'].fillna((new_merged_df['credit_score'].mean()), inplace=True)
```

```python
new_merged_df['prior_bi'].fillna((new_merged_df['prior_bi'].mean()), inplace=True)
```

```python
new_merged_df['time_w_carr'].fillna((new_merged_df['time_w_carr'].mean()), inplace=True)
```

```python
new_merged_df.isnull().values.any()
```

```python
#Removing 'curnt_insurer' column since there are lots of number of Missing values.Removing 'min_age','hoh_age','max_age','min_mon_lic','hoh_mon_lic','max_mon_lic','cnt_majr_viol','cnt_minr_viol','curnt_bi_low'
new_merged1_df=new_merged_df.drop(['curnt_insurer','min_age','hoh_age','max_age','min_mon_lic','hoh_mon_lic','max_mon_lic','cnt_majr_viol','cnt_minr_viol','curnt_bi_low'],axis=1)
```

```python
#Saving the cleaned and merged dataset

new_merged1_df.to_csv('Resources/merged_data.csv',index=False)
```
